#### <span style="color:#430600"><b>Forest cover analysis</b></span>

Using the above cubic model, let us investigate further details 
about deforestation in the Philippines.  The model makes it possible to 
address queries such as:

<B><span style="color:#3306B0">1. Estimate the amount of remaining 
forest cover in any specific year:</span></B>
Our original dataset is very sparse and contains only 6 data points 
between the years 1920 and 2005. From a practical standpoint, it may be 
of interest to know how serious the situation was in some other year, 
for example, in 1960.  We can easily estimate (the approximate) value 
from our cubic model.  How?  Just plug in $x=1960-1920=40$, and get
$$
  f(40) = 0.00021189\times 40^3 - 0.0252\times 40^2 + 0.1570\times 40 + 62.5244 
    = 41.98\%
$$
> In 1960 the forest cover that remained in the Philippines 
was approximately $42\%$ of the total land area.


<B><span style="color:#3306B0">2. Estimate <U>the rate</U> at which forest cover was being lost in any specific year:</span></B>
The rate of loss is a very important indicator, as it offers insights 
into how quickly action must be taken to address the problem.  Again, 
the cubic model makes this very easy to estimate, because the rate 
of loss is simply its first derivative:
$$
\begin{eqnarray*}
  f^\prime(x) & = & 3(0.00021189) x^2 - 2(0.0252) x + 0.1570 \\
   & = &  0.00063567 x^2 - 0.0504 x + 0.1570
\end{eqnarray*}
$$

Given below is a Python code segment that plots this function.  The 
plot shows that the derivative is negative for most of the period between 
1920 and 2005.  Of course, that is not surprising when you look at the 
earlier graphs of $f(x)$.  Can you explain why the derivative is mostly 
negative?

To gain a clearer understanding, observe that in 1980 (which 
corresponds to $x=60$), the value of $f^\prime \approx -0.59$ \% 
per year.  What does that mean?

> Since the derivative measures the rate of change, it means in 1980 the forest cover in the Philippines was decreasing at a rate of $0.59\%$ per year.  If this rate were constant, we could estimate that forests would cover $5.9\%$ less land area after $10$ years. Note, however, that this rate is not constant (see graph below).


<B><span style="color:#3306B0">3. In which year was the rate of loss 
the highest:</span></B>
A quick look at the graph of $f^\prime$ suggests the loss rate was 
highest around $x=40$, i.e., year=1960. However, let us use calculus and 
obtain a more precise estimate.  To state the problem in calculus 
terms:

> Where does $f^\prime$ attain its minimum value?

Since $f^\prime$ is continuous on a closed interval, calculus theory 
guarantees it will attain an absolute minimum at one of its critical 
points, or at an endpoint.  How do we find the critical points of a 
function?  That's easy: Just find the $x$ values where its derivative 
is 0 (or undefined).  In the case of $f^\prime$ we get
$$
 (f^\prime)^\prime = 2(0.00063567) x - 0.0504 = 0 
~ \Rightarrow ~ x = \frac{0.0504}{2(0.00063567)} = 39.64
$$
We leave it as an exercise for the student to show that $f^\prime$ 
attains its absolute minimum at this value of $x$.  Of course, from a 
practical standpoint, this value is very close to the year 1960, as 
we obseved from the graph of $f^\prime$.

```python
# Use Python to differentiate the polynomial
derivative = polynomial.deriv()

# Generate x values for plotting the derivative
x_smooth = np.linspace(min(x), max(x), 500)
y_deriv = derivative(x_smooth)

# Plot the derivative
plt.plot(x_smooth, y_deriv, color='green', label='Derivative of cubic')

# Add labels, legend, and title
plt.xlabel('Years since 1920')
plt.ylabel('Derivative of forest cover (% per year)')
plt.title('Rate of change of forest cover')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

Let's start with some "warmup" questions:

improving? If so, how? <br>

If so, in what way?

The information contained in this graph is insightful.  But, certain 
details of practical interest remain unclear.  For instance, can we 
estimate the total, or aggregate amount of forest cover lost between 
2004-2013, or over any specific time period? How much smaller was the 
aggregate loss in later years compared to that of earlier years?  In 
calculus language, we can find the net change in a quantity by 
integrating its rate of change.  Thus, we have:

- $\int_0^9 f(x) dx=$ total forest cover lost between 2004-2013
- $\int_0^4 f(x) dx=$ forest cover lost in 1st half (2004-2008)
- $\int_5^9 f(x) dx=$ forest cover lost in 2nd half (2009-2013)

Here $f(x)$ denotes the rate of deforestation, as seen in the 
graph, and $x$ denotes the number of years since 2004.  We can estimate 
the value of these intergrals using Riemann sums.  For example, a 
rough approximation for $\int_0^9 f(x) dx$ using left end-points 
with 9 rectangles is
$$
\begin{eqnarray*}
  \int_0^9 f(x) dx & \approx & (27772 + 19014 + 14286 + 11651 + 12911 + 7464 + 7000 + 6418 + 4571)  \\
  & = & 111,087 ~~ \mbox{sq.\ km.}
\end{eqnarray*}
$$

Similarly, using left end-points with 4 rectangles we get the 
approximations
- $\int_0^4 f(x) dx \approx 72,723 ~~ \mbox{sq.\ km.}$
- $\int_5^9 f(x) dx \approx 25,453 ~~ \mbox{sq.\ km.}$

As you're aware, it is possible to get somewhat better 
approximations using mid-points.  We leave it  as an exercise for 
the reader to show that mid-point sums with 4 rectangles give: 
$\int_0^4 f(x) dx \approx 65,293 ~~ \mbox{sq.\ km.}$, and 
$\int_5^9 f(x) dx \approx 24,643 ~~ \mbox{sq.\ km.}$

> From these results we note that the net loss in forest cover during 2009-2013 was less than half the amount lost between 2004-2008.

Thus, the situation is improving in the sense that Brazil has lost 
less forest land in recent years than in the past.  However, the fact 
remains that forest areas continue to shrink with each passing 
year, and deforestation remains a problem in Brazil.

### 3. Case study: China

In this study we approach deforestation from a different 
perspective. Farming and agricultural practices are known to contribute in 
significant ways to the loss of forest cover.  A useful quantitative 
indicator of the growth in agricultural production in a country is to 
look at the use of feed grains in that country.  

The Execl file 
`Verzosa_feedgraindata.xlsx` contains data on the amount of feed grain 
used in China and in the United States between the years 1960 and 
2012. (See data source in [Verzosa (2015): Reading the World with Calculus](https://doi.org/10.1080/10511970.2014.978983).</span>)
We used <A HREF="https://chat.openai.com" target="_blank">ChatGPT</A> 
to produce the following code to read and plot the data


```python
# Load the needed Python libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_name = "Verzosa_feedgraindata.xlsx"

# Specify header in row 3 (0-indexed row 2) and skip the first 5 rows
# Read 53 rows, as there is more text content after the rows of data.
data = pd.read_excel(file_name, header=2, skiprows=[3, 4], nrows=53)  # Skips only rows 4 and 5 (0-indexed)

# Inspect the first few rows to confirm the data is read correctly
print(data.head())

# Extract relevant columns
year = data['Year']-1960
china = data['China']

# Plot China vs Year
plt.figure(figsize=(10, 6))
plt.scatter(year, china, marker='o', linestyle='-', color='blue', label='China')

# Add labels, title, and legend
plt.xlabel('Years since 1960')
plt.ylabel('Feed grain use (million tons)')
plt.title('Feed grain use in China')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

Observe the graph is generally trending upwards, and can be approximated 
using a quadratic model, which will make the data easier to analyze. Again, 
we can find the best quadratic model using Python

```python
import numpy as np
# Use the numpy function "polyfit" to find cubic polynomial (degree 2)
coefficients = np.polyfit(year, china, deg=2)

# Display the polynomial coefficients
print("Quadratic Polynomial Coefficients (from highest to lowest degree):")
print(coefficients)
```

<B>Thus the quadratic model is</B>
$$
  f(x) = 0.04764 x^2 + 0.28424 x + 7.91709
$$
where $f(x)$ denotes the amount of feed grain used (in million 
tons) $x$ years after 1960. It is plotted below

```python
x = np.linspace(0, 55, 500)
plt.plot(x, 0.04764*x**2 + 0.28424*x + 7.91709)
plt.scatter(year, china, marker='o', linestyle='-', color='blue', label='China')
plt.show()
```

What insights can we gather from the graph of the quadratic model?  In 
particular, let's take a closer look the rate of change in feed grain use: 
- What does its sign tell us?
- Is its magnitude increasing?  What does that tell us?
- In what year between $[1960, 2012]$ is the rate of increase maximum?

Agricultural production, particularly the raising and feeding of cattle for humans to eat, is among the largest contributing factors to 
deforestation.  (See [How animal agriculture is accelerating global deforestation](https://earth.org/how-animal-agriculture-is-accelerating-global-deforestation).)  Thus, the rate of increase in the use of feed grains serves as a 
reasonable proxy for the rate of loss of forest lands.
