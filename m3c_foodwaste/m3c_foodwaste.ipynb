According to the UN, the world currently produces enough food 
to feed everyone on the planet.  Yet, every day many millions 
of people are on the brink of starvation and/or malnutrition due 
to lack of access to food.  This paradoxical situation is caused 
by several intertwining factors, one of which is the enormous 
scale of wasted food in the world.  In a 2011 report, the Food and 
Agriculture Organization of the UN states that "roughly one-third 
of food produced for human consumption is lost or wasted globally, 
which amounts to 1.3 billion tons per year."

The tragic consequences of wasted food are that it contributes not 
only  to global hunger, but also increases a host of unnecessary 
environmental costs associated with producing, transporting, 
storing and disposing it.  Although food waste occurs in both 
high- and low-income countries, there are differences in sources 
and causes.  In general, waste occurs at various points 
throughout the food supply chain, from production to 
consumption.  In the United States, the Environmental Protection 
Agency estimates food is the largest single item of "trash" that 
enters landfills and incinerators.

In this lab we construct models to study the amount of food wasted 
at the consumption end, and explore ways in which wasted food can

## <span style="color:#336630">Model development</span>

made; identify the variables used;  construct the math model;

### 1. Defining the problem

The goal of our problem may be stated as follows:

> We want to create a model to determine whether the amount of food wasted in any given state in the U.S. can be used to feed its food-insecure population.  The state of Texas will be used as example to demonstrate how the model works.

### 2. Strategy and assumptions


Before we begin, it is interesting to note that the winning 
teams arrived at completely different conclusions about Texas -- 
one team concluded the amount of wasted food is **not** 
sufficient to feed its food-insecure population, while another 
team concluded the opposite!  Yet, both teams were "right" and 
won prizes!  How can that happen?

A closer examination of the modeling strategy of these two teams 
reveals there were significant differences in the assumptions they 
made.  For example, food waste occurs at many different levels: 
individual households; supermarkets; cafes & restaurants; in 
warehouses; during transportation; in agriculture & harvesting, 
etc.  There are no clear guidelines on which type of food waste 
can be reused to feed people in need.  Teams 
made varying (but reasonable) assumptions on such questions, 
resulting in conclusions that were sometimes widely different.

The overall approach consists of estimating the 
amount (or dollar value) of food wasted in the state/province 
in one year, 
and comparing it with the amount of food needed to feed its 
food-insecure population for that year.  The estimate of food waste 
tends to differ significantly based on the assumptions 
made.  We will formulate a metric for food waste similar to that 
developed by Teams \#10494 and \#10917. The strategy involves 
making the following key assumptions

 1. All food wasted at any point in the production to consumption process will be considered usable for feeding the food-insecure.
 2. All wasted food can be transported and distributed to those in need without further waste, loss or damage.
 3. The dollar value of a food item is a reasonable proxy for the amount and/or quantity and/or calories of that item.
 4. All food-insecure individuals have the same food needs every day, and for the entire year.  Thus, we treat the food needs of children and adults the same.
 5. A state's cash receipts are a reasonable representation of all food purchased for consumption in that state.
 6. Average food waste percentages across North America are representative of the actual food waste in any specific state in the U.S.

While some of these assumptions may be hard to justify in a rigorous 
way (e.g., 1, 2 and 4), they offer a reasonable starting point for 
a process of iterative model development.  Thus, we will use them 
to build a first model.
 
 
 ### 3. The model and its variables

In order to determine whether there is enough food waste in a state to 
fulfill the needs of the food-insecure, we will formulate separate 
metrics for the annual waste and the annual needs, in units of US dollars ($W$ and $N$, 
respectively).  Then, we can define the net surplus as: $S = W-N$.  If 
$S$ is positive, then the food waste is sufficient to meet the 
needs of the food-insecure.  If $S$ is negative, then there is a 
deficit, which would mean the food waste is not sufficient 
for meeting the needs.  For practical implementation, we will make 
the following additional refinements, which are based on how the 
food data are sourced and organized:

- All food items will be classified into one of 5 categories: Protein, produce (fruits, vegetables, etc), grain, dairy, oil.
- Wasted food will be divided into two groups: Production waste, and consumption waste.

The "Venn diagram" style sketch below is helpful for conceptualizing 
the processes we will consider in our model.  The outermost 
region (light blue) represents the totality of food produced, 
including food wasted in production.  The inner grey region 
is the subset of food purchased for consumption, which is assumed 
to correspond to the cash receipts.  The innermost subset shown 
in white represents the food that is actually consumed.  In other 
words, everything outside the white region is wasted food, either 
in the production/distribution stage, or in the consumption 
stage.  Recall, these are all in units of US dollars.

#### Metric for wasted food

General data about the percentage of food loss/waste in North America 
are available in the file 
[Texas_food_data.xlsx](https://m3challenge.siam.org/wp-content/uploads/Texas_food_data.xlsx).  In addition, the file contains data on 
the annual cash receipts in the state of Texas for all major food categories.  Using 
these data, we can formulate the annual food waste metric as follows <BR>&nbsp;

<span style="color:#830600">
$$
\begin{eqnarray*}
  W & = & \mbox{consumption waste} + \mbox{production waste} \\
  \mbox{consumption waste} & = &(\mbox{cash receipts}) \times 
      \left(\frac{\mbox{\% consumption waste}}{100}\right) ~~~~~~~~~~~ (1)
\end{eqnarray*}
$$
</span>

To compute production waste we use an indirect method, because 
the data only gives percentage waste, without any baseline 
figure.  Accordingly, we use the following two expressions 
and solve for production waste

$$
\begin{eqnarray*}
  \mbox{production waste} & = & (\mbox{total production})\cdot 
    (\mbox{\% production waste}/100)  \\
  \mbox{total production} & = & (\mbox{production waste}) +
      (\mbox{cash receipts})
\end{eqnarray*}
$$

The result is

<span style="color:#830600">
$$
  \mbox{production waste} = 
      \frac{(\mbox{cash receipts})\cdot(\mbox{\% production waste}/100)}
      {1 - (\mbox{\% production waste}/100)} ~~~~~~~~~~~ (2)
$$
</span>

Equations (1)-(2) define $W$ in terms of data that is readily 
available to us.  We note, however, that the waste metric will be 
computed individually for each food typecash receipts, and the results 
added to estimate the overall value of $W$.  This is because 
the data indicates high variability in cash receipts and waste 
percentage across different food groups.


#### Metric for food needs

The overall approach for computing the annual needs metric 
in a state is as follows <BR>&nbsp;

<span style="color:#830600">
$$
  N = (\mbox{\# of food insecure persons in the state}) \times 
  (\mbox{annual cost of feeding 1 person})
  ~~~~~~~~~~~ (3)
$$
</span>

For the state of Texas, data in the file [Texas_food_data.xlsx](https://m3challenge.siam.org/wp-content/uploads/Texas_food_data.xlsx) 
indicates the number of food insecure persons in 2015 was $4.32$ 
million and the average cost of a meal was $\$2.59$.  If we assume 
each food insecure person requires $3$ meals per day, for $365$ 
days a year, the annual cost of feeding 1 person 
$= \$2.59\times 3\times 365 = \$2,836.05$.  Therefore, for Texas

$$
  N = \$2,836.05 \times 4.32 \mbox{ million} = \$12,252 \mbox{ million}
$$

However, this total dollar figure is not very useful for making 
comparisons with the amount of wasted food, since there are 
significant disparities between needs and waste for different food 
groups.  What we would like is a breakdown of $N$ across the 
food categories, e.g., $N_\mbox{protein}, ~ N_\mbox{produce},~\ldots$ 
etc.  Then we can determine whether the amount of wasted proteins 
is sufficient to meet the protein needs, and so on.

In order to determine an appropriate breakdown of $N$, we consider 
average food intake guidelines recommended by public health 
agencies.  Team \#10494 computed the following daily intake 
percentages of different food groups, 
based on guidelines published in a document by the 
US Department of Agriculture


| Food category | Percent of diet |
| :-------------| :---------------|
| protein  | 8.6 % |
| fruit, vegetables, roots, tubers | 51.4 % |
| grain | 4.3 % |
| dairy | 34.3 % |
| oil | 1.4 % |

We will split $N$ into different food groups according to the 
proportions in this table.  For example, 
$N_\mbox{protein}= \$12,252 \times 0.086 = \$1,053.7 
\mbox{ million}$.  Proceeding in this way, we get the following augmented 
table

| Food category | Percent of diet &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; | $N_\mbox{category}$ (million \$) |
| :-------------| :---------------| :-------------------|
| protein  | 8.6 % | 1,053.7 |
| produce | 51.4 % | 6,297.5 |
| grain | 4.3 % |  526.8 |
| dairy | 34.3 % | 4,202.4 |
| oil | 1.4 % | 171.6 |

The values of $N_\mbox{category}$ represent the annual needs for that 
category of food in the state of Texas.  We can now compute the 
surplus or deficit of wasted food.


#### Surplus/deficit computation for Texas

Using data in the file [Texas_food_data.xlsx](https://m3challenge.siam.org/wp-content/uploads/Texas_food_data.xlsx) 
we will compute $W_\mbox{category}$ for each of the 7 food 
waste categories listed: (1) `cereals`, (2) `roots & 
tubers`, (3) `oilseeds & pulses`, (4) `fruit & vegetables`, (5) `meat`, 
(6) `fish & seafood`, 
(7) `milk`.  For the purpose of determining surplus or deficit, 
we will match these categories with the five in the food needs metric 
by consolidating as follows:

- protein = `meat` + `fish & seafood`
- produce = `fruit & vegetables` + `roots & tubers`
- grain = `cereals`
- dairy = `milk`
- oil = `oilseeds & pulses`

It is straightforward enough to use a hand calculator for computing 
the waste metric for each category, although the data input may be 
cumbersome.  We prefer to use the Python code given below (which 
was largely generated by `ChatGPT`) to do the needed 
computations.  Note that the production waste percent shown 
in the code combines all the wastes that occurs prior to consumption.

```python
# Setup the input data
categories = ['meat', 'fish', 'fruit & veggies', 'roots & tubers', 'grain', 'dairy', 'oil']

# Data for cash receipts is in units of 1000 US dollars
cash_receipts_data = {
    'meat': [8470005, 1835520, 361309, 337269, 191218, 63803, 5643],
    'fish': [17722],
    'fruit & veggies': [1196184, 72500, 57293, 50430, 37848, 34884, 30319, 20664, 15392, 10873, 10543, 10193, 9240, 8580, 7043, 6932, 6707, 5585],
    'roots & tubers': [113522, 18368],
    'grain': [480609, 297693, 128209, 9618, 3929],
    'dairy': [1848140],
    'oil': [109663, 94020, 439982, 12894]
}

production_waste_percent = {
    'meat': 13.5,
    'fish': 27.5,
    'fruit & veggies': 38,
    'roots & tubers': 52,
    'grain': 10,
    'dairy': 5.7,
    'oil': 18
}

consumption_waste_percent = {
    'meat': 11,
    'fish': 33,
    'fruit & veggies': 28,
    'roots & tubers': 30,
    'grain': 27,
    'dairy': 15,
    'oil': 4
}

# Compute the total cash receipts for each category
cash_receipts = {category: sum(values) for category, values in cash_receipts_data.items()}

# Function to compute the waste metric
def compute_waste_metric(cash_receipts, production_waste_pct, consumption_waste_pct):
    production_waste = (cash_receipts * (production_waste_pct / 100)) / (1 - (production_waste_pct / 100))
    consumption_waste = cash_receipts * (consumption_waste_pct / 100)
    total_waste = production_waste + consumption_waste
    return total_waste

# Compute the waste metric for each category
waste_metrics = {}

for category in categories:
    cash = cash_receipts[category]
    prod_waste_pct = production_waste_percent[category]
    cons_waste_pct = consumption_waste_percent[category]
    
    waste = compute_waste_metric(cash, prod_waste_pct, cons_waste_pct)
    waste_metrics[category] = waste

# Print the waste metrics
print(f"Estimated food wastes per year in the state of Texas:\n")
for category, waste in waste_metrics.items():
    print(f"Waste metric for {category.capitalize()} = {waste/1000:,.2f} million USD")

print(f"\nCombined categories for later use:")
print(f"Waste metric for protein = Meat+Fish = \
{(waste_metrics['meat']+waste_metrics['fish'])/1000:,.2f} million USD")
print(f"Waste metric for produce = fruit+veggies+roots+tubers = \
{(waste_metrics['fruit & veggies']+waste_metrics['roots & tubers'])/1000:,.2f} million USD")
```

As always, the key question that begs to be addressed is, 
how reliable are these conclusions?  Should we believe these 
results?  To answer these questions, we first note that our 
model in this situation is just a set of simple algebraic calculations 
based on a bunch of assumptions.  The primary source of uncertainty 
in the results is due to uncertainties in the assumptions
and in the input data.

As discussed in the conclusions above, we assumed all food insecure 
individuals have the same needs year round.  While this may 
suffice for a rough approximation, it clearly overestimates the 
value of $N$ for all food categories.  A more accurate estimate 
of $N$ could well show there is a surplus of wasted food in 
some other categories as well (e.g., oil and grain).  Another assumption 
we made is that all wasted food, at any point in the 
production to consumption cycle, can be used to fulfill 
the needs of the food insecure.  But, this is clearly not true 
in reality for many reasons, including the perishable nature of some 
food items, and the logistics of procuring and transporting 
them on time.  Making real-world refinements to this assumption will 
result in lower values of $W$, epecially in categories such as 
produce, meat and fish.

To summarize, the particular choice of assumptions we've made 
play a crucial role in shaping the results and our conclusions.  In 
an iterative process of model development, our next step would 
be to refine these assumptions and incorporate them into an 
improved model.

To explore the effect of uncertainties in the input data, we 
will carry out some sensitivity studies in which we will perturb 
the input values by a small amount and observe the effect on 
$W$.  This is done in the Python code below.  We don't need to 
do this for $N$, since its behavior is linear with respect to 
perturbations in either of its inputs (number of food insecure 
people, cost of feeding per person).  That means, an $x\%$ change in 
either of its inputs will produce an $x\%$ change in $N$.

```python
# Load the needed libraries & assign nicknames
import numpy as np

# Define a function to compute the waste metric
def compute_waste_metric(cash_receipts, production_waste_pct, consumption_waste_pct):
    production_waste = (cash_receipts * (production_waste_pct / 100)) / (1 - (production_waste_pct / 100))
    consumption_waste = cash_receipts * (consumption_waste_pct / 100)
    total_waste = production_waste + consumption_waste
    return total_waste

# Define a function to perturb the inputs and compute sensitivity
def sensitivity_analysis(cash_receipts, production_waste_pct, consumption_waste_pct, perturbation=0.01):
    original_waste = compute_waste_metric(cash_receipts, production_waste_pct, consumption_waste_pct)
    
    # Perturb cash receipts
    perturbed_cash = cash_receipts * (1 + perturbation)
    perturbed_waste_cash = compute_waste_metric(perturbed_cash, production_waste_pct, consumption_waste_pct)
    
    # Perturb production waste percentage
    perturbed_prod_pct = production_waste_pct * (1 + perturbation)
    perturbed_waste_prod = compute_waste_metric(cash_receipts, perturbed_prod_pct, consumption_waste_pct)
    
    # Perturb consumption waste percentage
    perturbed_cons_pct = consumption_waste_pct * (1 + perturbation)
    perturbed_waste_cons = compute_waste_metric(cash_receipts, production_waste_pct, perturbed_cons_pct)
    
    # Compute percentage changes
    sensitivity_cash = (perturbed_waste_cash - original_waste) / original_waste * 100
    sensitivity_prod = (perturbed_waste_prod - original_waste) / original_waste * 100
    sensitivity_cons = (perturbed_waste_cons - original_waste) / original_waste * 100
    
    return sensitivity_cash, sensitivity_prod, sensitivity_cons


# Perform sensitivity computations for specified perturbation percent
perturbation = 0.1  # Example perturbation of 10%

for category in categories:
    cash = cash_receipts[category]
    prod_waste_pct = production_waste_percent[category]
    cons_waste_pct = consumption_waste_percent[category]

    sensitivity_cash, sensitivity_prod, sensitivity_cons = sensitivity_analysis(cash, prod_waste_pct, cons_waste_pct, perturbation)

    print(f"Sensitivity of '{category}' waste metric to 10% perturbation:")
    print(f"  Cash receipts: {sensitivity_cash:.2f}%")
    print(f"  Production waste percentage: {sensitivity_prod:.2f}%")
    print(f"  Consumption waste percentage: {sensitivity_cons:.2f}%")
    print()
```

#### Interpretation of sensitivity results

Since cash receipts are linearly related to $W$, we are not 
surprised to see that a $10\%$ change in its value causes a 
$10\%$ change in $W$ for all food categories.  As for 
other observations, it appears that changes in production waste 
cause larger changes in $W$ than changes in consumption 
waste, for most food categories.  In fact, a $10\%$ change in 
consumption waste percentage causes $W$ to change by less than 
$5\%$ in all categories, except 'grain' and 'dairy'.  Overall, these 
results suggest it is important to accurately estimate the values 
of (at least) the cash receipts and production waste percentage 
for all food categories, since they have a significant impact on 
the value of $W$.  Likewise, since the value of $N$ linearly 
depends on the total number of food insecure persons and the 
per person feeding cost, those variables are important to accurately 
estimate as well.
