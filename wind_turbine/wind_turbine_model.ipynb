## <span style="color:#336630">Model development</span> 

We will follow best practices guidelines in math modeling, 
which typically require 
addressing each of the following components in the course of model 
development: clearly define the problem; state any assumptions 
made; identify the variables used;  construct the math model; 
analyze and assess the solutions.

### 1. Defining the problem

The goal of our problem may be stated as follows:

> Can the Earlham College campus meet its annual electricity needs by installing a wind turbine on campus? To address this question, we will:
> 1. Develop a physics-based model for the power output of a wind turbine.
> 2. Use the model to estimate annual energy production.
> 3. Compare that estimate with Earlham College’s annual electricity consumption.
> 4. Assess whether the turbine could fully meet, partially meet, or fall well short of campus needs.

We have the following known or estimable quantities: 
- Typical wind speeds at the install location
- Physical characteristics of the proposed turbine (blade radius, efficiency)
- Annual electricity consumption of the campus

The unknown quantities, or outputs of interest are:
- Power output as a function of wind speed & other variables
- Total annual electrical energy generated by the turbine

### 2. Strategy and assumptions

From the problem statement and the desired end-goals, it is 
clear that we must model the turbine's power output as a 
function of wind speed.  The model will also necessarily 
depend on various parameters specific to the turbine being 
installed (e.g., it's size, mechanical efficiency, generator 
details, etc.).  Other information we will need to obtain 
or estimate:

- meteorological data on wind speeds at the install site.
- data on Earlham College’s annual electricity consumption.

Once we have a physical model for the power output as a 
function of wind speed, we can use a summation or integration strategy 
to estimate the total power generated in a full year, based on 
given wind speed data.  In the next section, we develop a simple 
physics-based model for the turbine, and show how to use it to 
estimate annual power output.  For clarity, we also discuss 
our assumptions in that section, in conjunction with model 
development.

### 3. The model and its variables

A quick briefing from the literature indicates that a wind turbine 
produces energy as follows:
<UL>
  <LI>It converts the wind’s kinetic energy into rotational KE in the turbine.
  <LI>The rotational kinetic energy is converted into electrical energy by a generator.
  <LI>The amount of energy converted depends primarily on wind speed and the swept area of the turbine blades.
  <LI>Efficiency losses occur at various internal conversion stages 
      in the turbine.
</UL>

The usual way of thinking (and modeling) the situation is in 
terms of power, which is simply the rate of energy production 
per unit time.  Accordingly, let $P_w$ denote the wind power 
that the turbine receives, and $P_e$ the electric power it 
outputs.  It is extremely helpful to keep careful 
track of units when developing math models.  We will use metric units 
for all quantities here.  In this system, the units of power 
are `joules/second`, more commonly known as `watts`.

<IMG SRC="./wind_power_sketch.png" width="350"> <BR> 
  
It is theoretically possible, from a 
simplified 1-D form of energy conservation laws, 
to estimate <BR><BR>
$~~~~~~~~~~~~\displaystyle P_w = \frac{1}{2} \rho A u^3 ~~~~~~\mbox{and}~~~~
\displaystyle P_e = C P_w$ <BR><BR>
where $\rho=$ air density ($kg/m^3$), $A=$ area swept by the 
turbine blades ($m^2$), $u=$ wind speed ($m/s$), 
$C=$ conversion coefficient that accounts for efficiency losses 
(unitless).  In terms of units, $P_w=(kg/m^3)(m^2)(m/s)^3 
= (kg\cdot m^2/s^2) \cdot (1/s) = \mbox{joules}/s$.

<span style="color:#303693">
Key assumptions underlying this model include:

1. Wind speed is approximately uniform across the turbine’s swept area.
2. Wind conditions are represented by an average or typical wind-speed profile.
3. Air density is constant.
4. Turbine efficiency is represented by a single coefficient.
</span>


Although this model is reasonable, the literature indicates 
it is difficult to estimate reliable values for its 
parameters.  Therefore, it is common practice to replace 
it with an empirical model, based on 
using a simple  **power curve**, such as the one shown below 
<IMG SRC="./wind_turbine_power_curve.png" ALIGN="CENTER" width="300"></IMG>

From this curve, the turbine's power output as a function of 
wind speed ($u$) has the form <BR><BR>
<span style="color:#933630">
$~~~~~~~~P_e(u)=\left\{\begin{array}{ll} a + b u^\alpha, & \mbox{ if } u_c \le u \le u_r  \\
            P_{er}, & \mbox{ if } u_r < u \le u_f \\ 
             0, & \mbox{ if } u<u_c \mbox{ or } u>u_f \end{array}\right.$ 
</span>

$~~~~u_c=$ cut-in wind speed, where electric power begins to be generated <BR>
$~~~~u_r=$ rated wind speed, above which power generation 
    remains constant<BR>
$~~~~u_f=$ furling wind speed, at which the turbine is shut off to prevent damage <BR>
$~~~~\alpha$ is a fitting parameter whose value is typically close to 2 <BR> 
$~~~~a$ and $b$ are found by curve-fitting once we know $u_c$, $u_r$, $\alpha$. <BR>
$~~~~P_{er}$ is the (constant) rated power output under optimal wind conditions.  <BR>
Numerical<SUP>&nbsp;</SUP>values of all these parameters depend on design, 
manufacture and installation properties of the wind turbine.

We will use the piecewise function $P_e(u)$ given above as our 
model for the turbine's electric power output.

<H4><FONT COLOR="#A30836">How to estimate total power output?</FONT></H4>

By integration, of course! But,let's be careful about an 
important and subtle issue: power vs energy.  Recall, power is the 
instantaneous rate of energy production (or consumption).  Its 
units are &rarr; *energy/second*.  The **total power output** over 
any time period is, therefore, equivalent to **energy**, and 
will have energy units.  

According to our problem statement, we 
want to estimate the annual energy production from the wind turbine.  This 
can be done using a 2-step process: (1) compute the power we expect 
the turbine to generate on average, say, ${P}_{Avg}$; (2) multiply 
${P}_{Avg}$ by the time period of interest.

To estimate ${P}_{Avg}$, we need one more piece of 
information, i.e., the distribution of actual wind speeds at the 
turbine location. This typically comes from historical 
meteorological data, in the form of a probability density 
function.  For example, a Weibull distribution is frequently 
used to model wind speed data.  

Conceptually, we can view the wind speed distribution 
as a histogram showing the relative frequency of occurrence of 
different wind speeds during some observation period. A Weibull 
pdf (probability density function) is a theoretical approximation 
of such a histogram, where the relative frequencies are 
interpreted as probability densities.

<TABLE ALIGN="LEFT" BORDER="0"  CELLSPACING="0"  CELLPADDING="0">
  <TR VALIGN="TOP">
    <TD WIDTH="50%">  
According to the Weibull pdf, the probability density corresponding 
to wind speed $u$ is given by: <BR> &nbsp;<BR>
<span style="color:#933630">
$\displaystyle \large f(u)=\frac{\frac{k}{c}(\frac{u}{c})^{k-1}}{e^{(\frac{u}{c})^k}}$  
</span>
<BR> &nbsp; <BR>
$k$ and $c$ are model parameters that depend on the 
median and variability in wind speeds. Note that the units of 
$f(u)$ are $s/m$.  (Probability densities always have inverse 
units of their dependent variable.) <BR><BR>
<BR><BR>
    </TD>
    <TD ALIGN="CENTER">
        <B>Weibull wind speed distribution</B> &nbsp; &nbsp; &nbsp; &nbsp; <BR>
      <IMG SRC="./wind_speed_distribution.png" width="300">
    </TD>
  </TR>
</TABLE>

From probability theory, the expected (or average) value of $P_e(u)$ 
is: <BR>
<span style="color:#933630">
$\displaystyle P_{Avg}=\int_0^\infty P_e(u) f(u) du$ 
</span><BR> 
where $P_e(u)$ and $f(u)$ are both given by expressions above.  For 
sanity check, the units of 
$P_{Avg} \sim \mbox{watts} \cdot (s/m) \cdot (m/s) = \mbox{watts}$, 
which is exactly what we want!  Thus, the total energy production 
is: 
$\displaystyle ~ E_{tot} = P_{Avg} \times$ (`# of hours of operation`). 
If we assume the turbine operates continuously through the 
entire year, the annual energy production is:  <BR>
<span style="color:#933630">
$\displaystyle ~ E_{tot} = P_{Avg} \times 24 \times 365$ 
</span>
&nbsp; (in units of watt-hours)  <BR>
Note that we could easily compute $E_{tot}$ in joules (simply 
multiply $P_{Avg}$ by \# of **seconds** of operation), but it 
is more common to express electric consumption in watt-hours or 
kilowatt-hours.

### 4. Results and model assessment

We will now carry out simulation studies using the model 
described above.  This will require computing the integral: 
$\displaystyle P_{Avg}=\int_0^\infty P_e(u) f(u) du$  <BR><BR>
with $~~~~~P_e(u)=\left\{\begin{array}{ll} a + b u^\alpha, & \mbox{ if } u_c \le u \le u_r  \\
            P_{er}, & \mbox{ if } u_r < u \le u_f \\ 
             0, & \mbox{ if } u<u_c \mbox{ or } u>u_f \end{array}\right.$
&nbsp; &nbsp; &nbsp; and 
$~~~~~~\displaystyle f(u)=\frac{\frac{k}{c}(\frac{u}{c})^{k-1}}{e^{(\frac{u}{c})^k}}$
<BR>


In order to proceed, we need to find 
numerical estimates for the various model parameters.  The wind 
turbine that Earlham College is investigating for installation 
is manufactured by General Electric, with model number 
GE 1.7-103.  Based on the model and installation, we found 
the following estimates for the parameters <BR>
$~~~u_c ~ \sim ~ 3\mbox{ to }4 ~m/s$; $~~u_r ~ \sim ~ 12\mbox{ to }14 ~m/s$ 
$~~u_f = 20 ~m/s$; <BR>
$~~P_{er} ~ \sim ~ 3000\mbox{ to }4000 ~kW$, depending on value or $u_r$; <BR>
$~~\alpha ~ \sim ~ 2\mbox{ to }2.3$; $~~k = 2.323$; $~~c=6.52 ~m/s$.  <BR>
$~~a$ and $b$ can be found if we know $u_c$, $u_r$, by curve-fitting. <BR>
It is left as an exercise for the reader to show that $b = P_{er} / (u_r^\alpha - u_c^\alpha)$ and $a=-b\cdot u_c^\alpha$.
<BR>

The Python code below computes the integral that estimates 
$P_{Avg}$.  The result is in kilowatts (kW) because the power curve 
values are given in kW.  Multiplying this by $24\times 365$ gives 
the electric energy generated per year in $kWh$, since we assume the 
turbine operates continuously through the entire year.  To run the code, 
click anywhere inside the code cell and select "Run" from the 
notebook menu above (near the top of this page).

```python
# The following Python code performs the integration to compute 
# average power, and then uses it to compute annual energy output.

import sympy as sp
u = sp.Symbol('u') # declare "u" as a variable
# Input the user parameter values
uc = 4            # cut-in speed (m/s)
ur = 12           # rated speed (m/s)
uf = 20           # furling speed (m/s)
per = 3157        # rated power output (kW)
alpha = 2.2       # exponent in power curve

c = 6.52          # speed parameter in Weibull distribution (m/s)
k = 2.323         # exponent in Weibull distribution
f = ( (k/c)*(u/c)**(k-1) ) / ( sp.exp((u/c)**k) )
a = (per*uc**alpha) / (uc**alpha - ur**alpha)
b = - per / (uc**alpha - ur**alpha)

# Integrate
otp = sp.integrate((a + b*u**alpha)*f, (u, uc, ur)) + sp.integrate(per*f, (u, ur, uf))
print(f'Estimated value of average power output = {otp:.1f} kW')
print(f'Estimated electric energy generated per year = {otp*24*365:.0f} kWh')
```

<H4><FONT COLOR="#A30836">Model assessment</FONT></H4>

As always, a key question of interest is, how reliable is our 
modeling strategy?  Firstly, we acknowledge that we have 
made some assumptions and approximations that can 
be improved.  But, the key goal of our analysis was to get a 
quick and dependable answer to the underlying question about 
meeting the electricity needs of the campus.  From that standpoint, 
the assumptions and approximations are reasonable.

Regarding the model components themselves &mdash; such as the power-curve 
method for estimating the turbine's output, and the probability 
based wind speed distributions &mdash; these approaches are standard 
in the literature.  However, an additional and important source of 
uncertainty arises from the estimated parameter values.  Conducting 
a parameter sensitivity analysis would therefore be instructive, 
as it would help identify which parameters, if any, 
exert a disproportionate influence on the model’s outputs.  Parameters 
with such influence are especially critical to estimate accurately, 
since small errors in their values can lead to significant 
changes in the results.

We leave it as an exercise for the reader to carry out the parameter 
sensitivity analysis.  This involves systematically tweaking each 
parameter value up (and down) by, say 20%, and observing the 
effect on the model's output.  The idea is to focus on the influence 
of one parameter at a time, keeping all the other parameters 
constant (see Exercise 1 below).
